```python
# Инструменты анализа данных
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Препроцессинг данных
from sklearn.preprocessing import StandardScaler, RobustScaler, SplineTransformer, PolynomialFeatures
from sklearn.decomposition import PCA, KernelPCA

# Система
import os

# Отображение
import warnings
warnings.filterwarnings('ignore')
sns.set_style('darkgrid')
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
%matplotlib inline
```

```python
# Пути
ROOT = os.getcwd()
TRAIN_DATASET = os.path.join(ROOT, '../data/train_AIC.csv')
TEST_DATASET = os.path.join(ROOT, '../data/test_AIC.csv')

# Загружаем датасеты
train_df = pd.read_csv(TRAIN_DATASET)
test_df = pd.read_csv(TEST_DATASET)
train_df = train_df.drop_duplicates()
gen_df = pd.concat([train_df.drop('y', axis=1), test_df], axis=0)
```

```python
for feature in train_df.columns[:-1]:
    plt.figure(figsize=(8, 3), dpi=300)
    sns.histplot(data=train_df, x=feature, hue='y', multiple='stack')
    plt.title(feature)
    plt.show()
```
