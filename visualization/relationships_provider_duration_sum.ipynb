```python
# Инструменты анализа данных
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Препроцессинг данных
from sklearn.preprocessing import StandardScaler, RobustScaler, SplineTransformer, PolynomialFeatures
from sklearn.decomposition import PCA, KernelPCA

# Система
import os

# Отображение
import warnings
warnings.filterwarnings('ignore')
sns.set_style('darkgrid')
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
%matplotlib inline
```

```python
# Пути
ROOT = os.getcwd()
TRAIN_DATASET = os.path.join(ROOT, '../data/train_AIC.csv')
TEST_DATASET = os.path.join(ROOT, '../data/test_AIC.csv')

# Загружаем датасеты
train_df = pd.read_csv(TRAIN_DATASET)
test_df = pd.read_csv(TEST_DATASET)
train_df = train_df.drop_duplicates()
gen_df = pd.concat([train_df.drop('y', axis=1), test_df], axis=0)
```

```python
plt.style.use('seaborn-poster')
fig = plt.figure(figsize = (20, 12))
ax = plt.axes(projection='3d')

x = train_df['Сумма']
y = train_df['Поставщик']
z = train_df['Длительность']

ax.scatter(x, y, z, c='r', s=20, alpha=0.3)
ax.set_xlabel('Сумма')
ax.set_ylabel('Поставщик')
ax.set_zlabel('Длительность')

plt.tight_layout()
plt.show()
```
